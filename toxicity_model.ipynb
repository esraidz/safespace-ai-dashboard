```python
import pandas as pd
import numpy as np

# 1. Aynı klasördeki CSV dosyasını okutuyoruz
df = pd.read_csv("labeled_data.csv")

# 2. Hız için tablodan sadece metin (tweet) ve radikallik (class) sütunlarını alıyoruz
df = df[['tweet', 'class']]

# 3. Modelimizi STK/Marka senaryosuna bağlamak için mock hashtagler üretiyoruz
markalar = ['#MarkaKrizi', '#BoykotX', '#KampanyaY', '#MusteriSikayeti']
df['aranan_hashtag'] = np.random.choice(markalar, size=len(df))

# 4. Verinin başarıyla yüklendiğini teyit etmek için ilk 5 satırı yazdırıyoruz
print("Veri başarıyla yüklendi! İlk 5 satır:")
display(df.head())

# 5. Ekip 2'nin arayüz için kullanacağı sahte veriyi aynı klasöre çıkarıyoruz
df.head(50).to_csv("arayuz_ekibi_icin_ornek.csv", index=False)
print("\n'arayuz_ekibi_icin_ornek.csv' dosyası klasöre kaydedildi. Ekip 2'ye iletebilirsin.")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression

# 1. ÇEVİRMEN (Kelimeleri bilgisayarın anlayacağı sayılara dönüştürür)
# max_features=5000: Sadece en sık kullanılan 5000 kelimeyi dikkate al ki hızlanalım.
print("1. Çevirmen hazırlanıyor...")
vectorizer = TfidfVectorizer(max_features=5000, stop_words='english')

# 2. BEYİN (Metinlerin sayisal karsiliklarina bakarak 0, 1, 2 olmayi ogrenen algoritma)
model = LogisticRegression(max_iter=1000)

# 3. VERİYİ DÖNÜŞTÜRME
print("2. Tweetler sayılara çevriliyor (Bu 10-15 saniye sürebilir)...")
X = vectorizer.fit_transform(df['tweet']) # Girdiler (Tweetler)
y = df['class'] # Çıktılar (0, 1, 2 etiketleri)

# 4. EĞİTİM (Makine Öğrenmesi tam olarak burada gerçekleşiyor!)
print("3. Yapay Zeka eğitiliyor, lütfen bekleyin...")
model.fit(X, y)
print("🚀 EĞİTİM TAMAMLANDI! Modelimiz artık metinleri anlayabiliyor.\n")

# 5. ARAYÜZ İÇİN KÖPRÜ (Bu fonksiyon Ekip 2'nin arayüzüne veri gönderecek)
def hashtag_analizi_yap(hedef_hashtag):
    # Tablodan sadece istenen hashtage ait tweetleri filtrele
    ilgili_tweetler = df[df['aranan_hashtag'] == hedef_hashtag]['tweet']
    
    if len(ilgili_tweetler) == 0:
        return {"hata": "Bu hashtage ait veri bulunamadı."}
    
    # Bu yeni tweetleri sayılara çevir
    yeni_sayilar = vectorizer.transform(ilgili_tweetler)
    
    # Eğittiğimiz beyinden tahmin etmesini iste
    tahminler = model.predict(yeni_sayilar)
    
    # Tahminlerin yüzdelerini hesapla
    toplam = len(tahminler)
    nefret_orani = (list(tahminler).count(0) / toplam) * 100
    saldirgan_orani = (list(tahminler).count(1) / toplam) * 100
    temiz_orani = (list(tahminler).count(2) / toplam) * 100
    
    return {
        "Hashtag": hedef_hashtag,
        "İncelenen Tweet Sayısı": toplam,
        "Nefret Söylemi / Radikal Oranı": f"%{nefret_orani:.1f}",
        "Saldırgan / Toksik Oranı": f"%{saldirgan_orani:.1f}",
        "Temiz Oran": f"%{temiz_orani:.1f}"
    }

# SİSTEMİ TEST EDELİM!
sonuc = hashtag_analizi_yap('#MarkaKrizi')
print("--- TEST SONUCU ---")
for anahtar, deger in sonuc.items():
    print(f"{anahtar}: {deger}")
```

```python
import joblib

# Çevirmeni (vectorizer) ve Beyni (model) bilgisayara kaydediyoruz
joblib.dump(vectorizer, 'cevirmen.pkl')
joblib.dump(model, 'yapay_zeka_modeli.pkl')

print("✅ Model ve Çevirmen başarıyla kaydedildi! Artık arayüze bağlanmaya hazır.")
```
